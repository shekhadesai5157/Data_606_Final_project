```python
# lets plot the distribution graphs i.e bar graphs and histograms
def hg(df, n, gpr):
    nunique = df.nunique()
   
    #lets pick coloumns which have 1-20 unique values.
    df = df[[i for i in df if nunique[i] > 1 and nunique[i] < 20]] 
    rows, cols = df.shape
    #gathering all the coloumns 
    cn = list(df)
    row = (cols + gpr - 1) // gpr
    
    plt.figure(num = None, figsize = (6 * gpr, 8 * row), dpi = 150, facecolor = 'w', edgecolor = 'k')
    
    for i in range(min(cols, n)):
        plt.subplot(row, gpr, i + 1)
        c_df = df.iloc[:, i]
        if (not np.issubdtype(type(c_df.iloc[0]), np.number)):
            vc = c_df.value_counts()
            vc.plot.bar()
        else:
            c_df.hist()
        plt.ylabel('all The counts')
        plt.xticks(rotation = 90)
        plt.title(f'{cn[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
```

```python
hg(df, 10, 3)
```

```python
# Correlation matrix
def cm(df, w):
    corr = df.corr()
    plt.figure(num=None, figsize=(w, w), dpi=350, facecolor='w', edgecolor='b')
    cmat = plt.matshow(corr, fignum = 1)
    
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(cmat)
    plt.title(f'Correlation Matrix ', fontsize=30)
    plt.show()
```

```python
cm(df, 20)
```

```python
# plotting the scatter and the density plots
def sm(df, ps, ts):
    df = df.select_dtypes(include =[np.number]) # we are only keeping the numerical coloumns 
    df = df[[col for col in df if df[col].nunique() > 1]] # lets omit the coloumns that dont have atleast one unique value
    cn = list(df)
    if len(cn) > 10: # Let's decrease the number of columns for kernel density charts' matrix inversion.
        cn = cn[:10]
    
    df = df[cn]
    Axis = pd.plotting.scatter_matrix(df, alpha=0.9, figsize=[ps, ps], diagonal='kde')
    cor = df.corr().values
    for i, j in zip(*plt.np.triu_indices_from(Axis, k = 1)):
        Axis[i, j].annotate('Crr coeff = %.3f' % cor[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=ts)
    
    plt.suptitle('Scatter and density plots ')
    plt.show()
```

```python
sm(df, 50, 30)
```
