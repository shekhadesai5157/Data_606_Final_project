```python
X = df_stage2.drop(['current_mental_disorder','employer_discussed_mental_health'], axis=1)
y = df_stage2['current_mental_disorder']
```

```python
df_stage2['current_mental_disorder'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("X_train dataset: ", X_train.shape)
print("y_train dataset: ", y_train.shape)
print("X_test dataset: ", X_test.shape)
print("y_test dataset: ", y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
# instantiate the DecisionTreeClassifier model with criterion entropy

dt_classifier = DecisionTreeClassifier(criterion='entropy', max_depth=8, random_state=8, splitter ='random')

# fit the model
dt_classifier.fit(X_train, y_train)
 
y_pred_en = dt_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

y_pred_train_en = dt_classifier.predict(X_test)
print('accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_train_en)*100))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

lr_model = LogisticRegression(solver='liblinear')
lr_model.fit(X_train, y_train)
predicted = lr_model.predict(X_test)
report = classification_report(y_test, predicted)
print(report)
```

```python
from sklearn.svm import SVC  
svc_model = SVC(kernel='linear') 
  
# fitting x samples and y classes 
svc_model.fit(X_train, y_train) 
```

```python
from sklearn.metrics import accuracy_score
y_pred = svc_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy:", accuracy)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
max_features_range = np.arange(1,6,1)
n_estimators_range = np.arange(10,210,10)
param_grid_meantal_tech = dict(max_features=max_features_range, n_estimators=n_estimators_range)
rf_classifier = RandomForestClassifier()
grid = GridSearchCV(estimator=rf_classifier, param_grid=param_grid_meantal_tech, cv=5)
grid.fit(X_train, y_train)
print("Optimal parameters:- %s accuracy score of:-  %0.2f"
      % (grid.best_params_, grid.best_score_))
```
